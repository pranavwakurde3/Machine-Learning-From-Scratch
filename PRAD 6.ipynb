```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# 1. Generate sample data with outliers
np.random.seed(42)
# Create inliers (normal data points) in two clusters
X_inliers = 0.3 * np.random.randn(100, 2)
X_inliers = np.r_[X_inliers + 2, X_inliers - 2]
# Create outliers (anomalous data points)
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))
# Combine data
X = np.r_[X_inliers, X_outliers]

# 2. Define the model
# The 'n_neighbors' parameter influences the "locality" of the assessment
# 'contamination' is the expected proportion of outliers in the dataset
clf = LocalOutlierFactor(n_neighbors=20, contamination=0.1)

# 3. Fit the model and predict outliers
# fit_predict returns -1 for outliers and +1 for inliers
y_pred = clf.fit_predict(X)
scores = clf.negative_outlier_factor_

# 4. Visualize the results
plt.figure(figsize=(10, 6))
# Plot inliers in blue
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], c='blue', label='Inliers')
# Plot outliers in red
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], c='red', label='Outliers')
plt.title("Outlier detection with Local Outlier Factor (LOF)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

# Print number of outliers detected
print(f"Number of outliers detected: {np.sum(y_pred == -1)}")
```

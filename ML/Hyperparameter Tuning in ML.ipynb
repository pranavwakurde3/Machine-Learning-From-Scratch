1) Manual Search HT

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data=load_iris()
x=data.data
y=data.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
for depth in [1,2,3,4,5]:
    model=DecisionTreeClassifier(max_depth=depth)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(f"depth:{depth},accuracy:{accuracy_score(y_test,y_pred)}")
```

2) GridSearch HT : Try ALL combinations of given hyperparameters.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
model = RandomForestClassifier()
```

```python
model.get_params()
```

```python
param_grid={
    'n_estimators':[50,100],
    'max_depth':[None,5,10]
}
```

```python
grid = GridSearchCV(model, param_grid, cv=5)
grid.fit(X, y)
```

```python
print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_) 
```

3) Random Search : Instead of checking all combinations,
                it randomly selects some combinations.
                Faster than Grid Search.

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
import numpy as np
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
model = RandomForestClassifier()
```

```python
model.get_params()
```

```python
param_dist={
    'n_estimators':np.arange(50,200),
    'max_depth':[None,5,10,20]
}
```

```python
random_search = RandomizedSearchCV(model,param_dist,n_iter=10,cv=5)
```

```python
random_search.fit(X, y)
```

```python
print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)
```

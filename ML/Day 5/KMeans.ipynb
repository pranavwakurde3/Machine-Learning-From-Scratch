```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
d1=np.random.normal(loc=[2,4],scale=1,size=(100,2))
d2=np.random.normal(loc=[4,8],scale=1,size=(100,2))
d3=np.random.normal(loc=[8,16],scale=1,size=(100,2))
data=np.vstack((d1,d2,d3))
```

```python
plt.scatter(data[:,0],data[:,1],s=50,color="red",label="actual")
plt.legend()
```

```python
model=KMeans(n_clusters=3,random_state=42)
```

```python
model.fit(data)
```

```python
plt.scatter(data[:,0],data[:,1],c=model.labels_,cmap="viridis",s=30)
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],marker="*",s=200,color="y",label="centroids")
plt.legend()
plt.title("Kmeans clustering")
plt.show()
```

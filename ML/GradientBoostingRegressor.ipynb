Gradient Boosting Regressor (GBR) is a powerful machine learning ensemble technique that predicts continuous target values by sequentially building, adding, and correcting errors (residuals) of shallow decision trees.

It can minimize the loss fuction line Mean Squared Error

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_digits
import numpy as np
```

```python
np.random_seeds=42
```

```python
X,y=load_digits(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
model=GradientBoostingClassifier(n_estimators=300,
                                 learning_rate=0.05,
                                 random_state=100,
                                 max_features=5)
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_pred)*100
acc
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test,y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("RBC - Actual vs Predicted")
plt.plot([min(y_test), max(y_test)], 
         [min(y_test), max(y_test)], color='red')
plt.show()
```

Random Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_diabetes
import numpy as np
```

```python
np.random_seed=42
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
x_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
model=GradientBoostingRegressor(loss='absolute_error',
                                learning_rate=0.1,
                                n_estimators=300,
                                max_depth=1,
                                max_features=5,
                                random_state=42)
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)**(1/2)
mse
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test,y_pred)
plt.xlabel('Actual value')
plt.ylabel('Predicted value')
plt.title("RBR - Actual vs Predicted")
plt.plot([min(y_test), max(y_test)], 
         [min(y_test), max(y_test)], color='red')
plt.show()
```

```python
import numpy as np
```

```python
X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])
```

```python
w=np.zeros(2)
b=0
lr=0.01
```

```python
for epochs in range(10):
  for i in range(len(X)):
    y_pred=np.dot(X[i],w)+b
    output=1 if y_pred > 0 else 0
    w+=lr*(y[i]-output)*X[i]
    b+=lr*(y[i]-output)
```

```python
print(w,b)
```

```python
#slicing[start:stop:step]
import numpy as np
arr=np.array([10,20,30,40,50,60])
print(arr[1:5])
print(arr[:4])
print(arr[::2])
print(arr[::-1])
```

```python
#flitering(Boolean Masking)
import numpy as np
arr=np.array([10,20,30,40,50,60])
print(arr[arr>25])
```

```python
#reshaping and manipulating arrays 
"""Syantax : reshape(row,coloumn) specify new shape
if dimension match"""
  
import numpy as np
arr=np.array([1,2,3,4,5,6])
re_shape=arr.reshape(2,3)
re_shape
```

```python
""".ravel() -> view
.flatten() -> copy"""

import numpy as np
arr=np.array([[1,2,3],[5,6,7]])
print(arr.ravel())
print(arr.flatten())
```

```python
#Advanced numpy
"""
np.insert(array,index,value,axis=None)
"""
import numpy as np
arr=np.array([10,20,30,40,50,60])
print(arr)
new_arr=np.insert(arr,2,18)
new_arr
```

```python
import numpy as np
arr=np.array([[1,2],[3,4]])
new_arr=np.insert(arr,1,[5,6],axis=1)
new_arr
```

```python
"""
np.append()
"""
import numpy as np
arr=np.array([10,20,30,40,50,60])
new_arr=np.append(arr,70)
new_arr
```

```python
"""
np.concatenate({array1,array2,axis=0})
axis=0 -> vertical stacking
axis=1 -> Horizanatal stacking
"""
import numpy as np
arr_1=np.array([1,2,3])
arr_2=np.array([4,5,6])
new_arr=np.concatenate((arr_1,arr_2))
new_arr
```

```python
"""
np.delete(array,index,axis=None)
"""
import numpy as np
arr=np.array([10,20,30,40,50])
print(arr)
new_arr=np.delete(arr,1)
new_arr
```

```python
import numpy as np
arr=np.array([[1,2],
             [3,4],
             [5,6]])
print(arr)
new_arr=np.delete(arr,1,axis=0)
new_arr
```

```python
"""
vstack(),hstack()
"""
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
print(np.vstack((arr1, arr2)))
print(np.hstack((arr1, arr2)))
```

```python
"""
np.split() -> equal
np.hsplit() ,np.vsplit()
"""
import numpy as np
arr=np.array([10,20,30,40,50,60])
print(np.split(arr,3))
```

```python
#Handling missing values
"""
1) np.isnan(array) -> datect missing value 
2) np.nan_to_num(array,nan=value) ->replace no
3) np.isinf()
"""
import numpy as np
arr=np.array([1,2,np.nan,4,np.nan])
print(np.isnan(arr))
print(np.nan==np.nan)
```

```python
import numpy as np
arr=np.array([1,2,np.nan,4,np.nan])
new_arr=np.nan_to_num(arr,nan=100)
new_arr
```

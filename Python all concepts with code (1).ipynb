```python
#Constructor 
class student:
    def __init__(self,roll_no,name):   #Invocate at the time of object creation.It is special method.
        self.roll_no=roll_no
        self.name=name
    def print_detail(self):
        print("roll_no=",self.roll_no)
        print("name=",self.name)
stu1=student(44,"Pranav")
stu1.print_detail()
```

```python
#1)Method Overloading- It means clling the subclass instead of parent class
class Bird():
    def sound(self):
        print("Bird make nice sound")
class crow(Bird):
    def sound(self):
        print("cow cow")
class Parrot(crow):
    def sound(self):
        print("sque sque")
Crow=crow()
parrot=Parrot()
Crow.sound()
parrot.sound()
```

```python
#Encapsulation - It means the binding the data and functions without showing to the user.
class BankAccount:
    def __init__(self,ac_no,balance):
        self.ac_no=ac_no
        self.__balance=balance #private varialbe
        
    def deposite(self,amount):
        self.__balance+= amount
        print("AC_NO:",self.ac_no)
        print("Deposit:",amount)
        print("New balance:",self.__balance)
        
    #def get_balance(self):
       #return self.__balance
        
account=BankAccount('123',5000)
account.deposite(5000)
#account.get_balance() 
```

```python
#Inheritance - It means the child class can access the parent class
'''Types-1)single
          2)multi-level
          3)multiple
          4)Hubrid
          5)Hirarchical '''

class Base():
    def ShowA(self):
        print("HI")
class Derived(Base):
    def ShowB(self):
        print("Byy")
#a1=Base()
b1=Derived()
b1.ShowA()
b1.ShowB()
```

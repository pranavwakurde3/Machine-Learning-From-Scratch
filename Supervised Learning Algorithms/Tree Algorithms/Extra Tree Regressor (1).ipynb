The Extra Trees Regressor (Extremely Randomized Trees) is an ensemble machine learning algorithm that fits multiple randomized decision trees on various sub-samples of a dataset, averaging their predictions to reduce over-fitting and improve accuracy.

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Generate sample data
X, y = make_regression(n_features=4, n_informative=2, random_state=0, shuffle=False)

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)

# Initialize and fit the model
model = ExtraTreesRegressor(n_estimators=100, random_state=0)

model.fit(X_train, y_train)

# Predict
predictions = model.predict(X_test)

print(predictions[:10])
```

**KNN is Supervised machine learning algorithm that makes prediction based on K closest data pounts in the dataset**

**It works by finding the "k" closest data points (neighbors) to a given input and makes a predictions based on the majority class (for classification) or the average value (for regression).**

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y =make_regression(n_samples=200,n_features=1,noise=20,random_state=42)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=KNeighborsRegressor(n_neighbors=5)
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mean_squared_error(y_test,y_pred)/10)
```

```python
print(r2_score(y_test,y_pred)*100)
```

```python
plt.scatter(x_test, y_test)
plt.scatter(x_test, y_pred, color='red')
plt.title("KNN Regression")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.show()
```

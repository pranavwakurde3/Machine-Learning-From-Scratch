```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X,y = make_moons(
    n_samples=300,
    noise=0.05,
    random_state=42
)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X[:,0],X[:,1])
plt.title("Original Dataset")
plt.show()
```

```python
dbscan = DBSCAN(
    eps=0.2,
    min_samples=5
)
```

```python
labels = dbscan.fit_predict(X)
labels
```

```python
plt.figure(figsize=(6,5))
plt.scatter(
    X[:,0],
    X[:,1],
    c=labels,
    cmap='rainbow'
)

plt.title("DBSCAN Clustering")
plt.show()
```

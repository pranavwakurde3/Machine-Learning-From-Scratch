**Bernoulli Naive Bayes is a Naive Bayes classifier used for binary features, where each feature represents the presence (1) or absence (0) of a particular attribute.**

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.feature_extraction.text import CountVectorizer

# Example text dataset
texts = [
    "free money offer",
    "win cash prize",
    "hello friend",
    "let us meet tomorrow"
]

labels = [1, 1, 0, 0]  # 1 = spam, 0 = not spam

# Convert text to binary features
vectorizer = CountVectorizer(binary=True)
X = vectorizer.fit_transform(texts)

# Create model
model = BernoulliNB()

# Train model
model.fit(X, labels)

# Predict
test = vectorizer.transform(["free prize"])
print(model.predict(test))
```

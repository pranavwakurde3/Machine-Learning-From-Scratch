**MLP Classifier (Multi-Layer Perceptron) is a neural network used for classification that consists of multiple layers of neurons to learn complex patterns.**

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score

# Create dataset
X, y = make_classification(n_samples=300, n_features=4, random_state=42)

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = MLPClassifier(
    hidden_layer_sizes=(100,50),
    activation='relu',
    max_iter=500,
    random_state=42
)

# Train
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```
